```python
### Scatter Plot/ Correlation
x = np.random.normal(100, 10, 200)
y = np.random.normal(90, 20, 200)

plt.scatter(x,y)
# different colors for the different variables

plt.xlabel('x')

plt.ylabel('y')

plt.show()
```

```python
### Linear Regression
x = np.random.normal(100, 10, 200)
y = np.random.normal(90, 20, 200)

coef = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef)

plt.plot(x,y, 'yo', x, poly1d_fn(x), '--k') # x, y = data, 'yo' = yellow circles, '--k' = black dashed line
plt.show()

# https://stackoverflow.com/questions/6148207/linear-regression-with-matplotlib-numpy
```

```python
### Box Plot
x = np.random.normal(100, 10, 200)
y = np.random.normal(90, 20, 200)

data = [x,y]

fig = plt.figure(figsize =(10, 5))

ax = fig.add_axes([0, 0, 1, 1])
 
bp = ax.boxplot(data)

ax.set_xticklabels(['x', 'y'])

plt.show()
# https://www.geeksforgeeks.org/box-plot-in-python-using-matplotlib/
```
